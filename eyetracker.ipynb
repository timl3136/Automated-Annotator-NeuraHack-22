```python
from pynput.keyboard import Key, Controller
import pyautogui, sys
```

```python
keyboard = Controller()
```

```python
try:
    while True:
        keyboard.press(Key.space)
        keyboard.release(Key.space)
        x, y = pyautogui.position()
        print(x,y)
except KeyboardInterrupt:
    print('Finished painting')
```
